```python
import time as t
def fib(n):
    if n==0 or n==1 :
        return 1 
    else:
        return fib(n-1)+fib(n-2)
start=t.time()
fib(30)     
print(f" time taken for executing this function without memoization : {t.time()-start}")    
def fib(n,d):
    if n in d :
        return d[n]
    else:
        d[n]=fib(n-1,d)+fib(n-2,d)
        return d[n]
start=t.time()    
d={0:1,1:1}
fib(30,d)
print(f"time taken for executing this function after memoization : {t.time()-start}")
    
```
