```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
data=pd.read_csv('/Users/rony/Documents/DJANGO/FastAPI/DEEP LEARNING/01_fundamentals/placement.csv')
data
```

```python
data1=data.iloc[:,:-1]
data2=data.iloc[:,(-1)]
data1,data2
X_train,X_test,y_train,y_test=train_test_split(data1,data2,test_size=0.33,random_state=42)
p=Perceptron()
p.fit(X_train,y_train)
y=p.predict(X_test)
y
plt.scatter(X_test.iloc[:,(0)],X_test.iloc[:,(1)],c=y)
w = p.coef_[0]
b = p.intercept_[0]

x1 = np.linspace(
    X_test.iloc[:,0].min(),
    X_test.iloc[:,0].max(),
    100
)

x2 = -(w[0]*x1 + b) / w[1]
plt.plot(x1,x2)
plt.xlabel("cgpa")
plt.ylabel("resume_score")
plt.title("Perceptron graph")
plt.show()
```
